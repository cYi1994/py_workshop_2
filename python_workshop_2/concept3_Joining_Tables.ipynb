```python
left = pd.DataFrame({'k1':[1,2,3,4],
                    'A':[12313,12455,34647,4313],
                    'B':[12345,123124,345346,45756]})
```

```python
left
```

```python
right = pd.DataFrame({'k1':[1,2,3,5],
                    'E':[12313,12455,34647,4313],
                    'F':[12345,123124,345346,45756]})
```

```python
right
```

```python
results = pd.merge(left,right, on='k1', how='inner')
```

```python
results
```

```python
results = pd.merge(left,right, on='k1', how='left')
results
```

```python
results = pd.merge(left,right, on='k1', how='right')
results
```

```python
results = pd.merge(left,right, on='k1', how='outer')
results
```
